```python
movie_correlation = user_movie_mat.corr(method='pearson',min_periods=80)
```

```python
movie_correlation
```

```python
my_ratings = pd.DataFrame({'title':['Show, The (1995)','Return of the Jedi (1983)'],'ratings':[5,1]})
```

```python
my_ratings
```

```python
my_ratings['title'][0]
```

```python
rec_list = pd.Series()

for i in range(0,2):
    similar_movie = movie_correlation[my_ratings['title'][i]]
    similar_movie = similar_movie.map(lambda x: x * my_ratings['ratings'][i])
    rec_list = rec_list.append(similar_movie)
```

```python
rec_list.sort_values(inplace=True,ascending=False)
```
